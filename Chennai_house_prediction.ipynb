```python
#No Of missing values is negligible.Hence drop those rows.

df.dropna(inplace=True)
df.isna().sum()
```

```python
#Cleaning the data

df['AREA'] = df['AREA'].str.lower()
df['AREA'].replace({'velchery':'velachery', 
                       'kknagar':'kk nagar', 
                       'tnagar':'t nagar', 
                       'chormpet':'chrompet', 
                       'chrompt':'chrompet', 
                       'chrmpet':'chrompet', 
                       'ana nagar':'anna nagar', 
                       'ann nagar':'anna nagar',
                       'karapakam':'karapakkam', 
                       'adyr':'adyar'}, inplace=True)
```

```python
df['SALE_COND'].replace({'Ab Normal':'AbNormal',
                           'PartiaLl':'Partial',
                           'Adj Land':'AdjLand',
                           'Normal Sale':'NormalSale',
                           'Partiall':'Partial'},inplace=True)
df['SALE_COND'] = df['SALE_COND'].str.lower()
```

```python
df['PARK_FACIL'].replace(['Noo'],'No',inplace=True)
df['PARK_FACIL'] = df['PARK_FACIL'].str.lower()
```

```python
df['UTILITY_AVAIL'].replace({'All Pub':'AllPub',
                              'NoSeWa':'NoSewr',
                              'NoSewr ':'NoSewr'},inplace=True)
df['UTILITY_AVAIL'] = df['UTILITY_AVAIL'].str.lower()
```

```python
df['BUILDTYPE'].replace({'Others':'Other',
                          'Comercial':'Commercial'},inplace=True)
df['BUILDTYPE'] = df['BUILDTYPE'].str.lower()
```

```python
df['STREET'].replace({'Pavd':'Paved',
                          'NoAccess':'No Access'},inplace=True)
df['STREET'] = df['STREET'].str.lower()
```

```python
df['MZZONE'] = df['MZZONE'].str.lower()
```

```python
df['DATE_SALE'] = pd.to_datetime(df['DATE_SALE'], format='%d-%m-%Y')
df['DATE_BUILD'] = pd.to_datetime(df['DATE_BUILD'], format='%d-%m-%Y')
df[['DATE_SALE','DATE_BUILD']]
```

```python
df['AGE_OF_PROPERTY'] = pd.DatetimeIndex(df['DATE_SALE'] ).year - pd.DatetimeIndex(df['DATE_BUILD']).year


df['TOTAL_PRICE'] = df['REG_FEE'] + df['COMMIS'] + df['SALES_PRICE']

df[['AGE_OF_PROPERTY','TOTAL_PRICE']]
```
